# Feature Engineering for Training dataset_train

We will be performing all the below steps in Feature Engineering

1. Missing values
2. Temporal variables
3. Categorical variables: remove rare labels
4. Standarise the values of the variables to the same range

```python
## Always remember there way always be a chance of data_train leakage so we need to split the data_train first and then apply feature
## Engineering
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(dataset_train,dataset_train['SalePrice'],test_size=0.1,random_state=0)
```

```python
X_test.shape, X_train.shape
```

## Missing Values

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing
features_nan=[feature for feature in dataset_train.columns if dataset_train[feature].isnull().sum()>1 and dataset_train[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(dataset_train[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(dataset_train,features_nan):
    data_train=dataset_train.copy()
    data_train[features_nan]=data_train[features_nan].fillna('Missing')
    return data_train

dataset_train=replace_cat_feature(dataset_train,features_nan)

dataset_train[features_nan].isnull().sum()
```

```python
dataset_train.head(50)
```

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in dataset_train.columns if dataset_train[feature].isnull().sum()>1 and dataset_train[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(dataset_train[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values

for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=dataset_train[feature].median()

    ## create a new feature to capture nan values
    dataset_train[feature+'nan']=np.where(dataset_train[feature].isnull(),1,0)
    dataset_train[feature].fillna(median_value,inplace=True)

dataset_train[numerical_with_nan].isnull().sum()

```

```python
dataset_train.head(50)
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:

    dataset_train[feature]=dataset_train['YrSold']-dataset_train[feature]
```

```python
dataset_train.head()
```

```python
dataset_train[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

## Numerical Variables

Since the numerical variables are Skewed, log normal distribution must be performed.

```python
dataset_train.head()
```

```python
import numpy as np
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset_train[feature] = np.log(dataset_train[feature])
```

```python
dataset_train.head()
```

# Handling Rare Categorical Feature

We will remove categorical variables that are present less than 1% of the observations.

```python
categorical_features = [feature for feature in dataset_train.columns if dataset_train[feature].dtypes =='O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp = dataset_train.groupby(feature)['SalePrice'].count()/len(dataset_train)
    temp_df = temp[temp>0.01].index
    dataset_train[feature] = np.where(dataset_train[feature].isin(temp_df), dataset_train[feature],'Rare_var')
```

```python
dataset_train.head(100)
```

## Feature Scaling

```python
for feature in categorical_features:
    labels_ordered=dataset_train.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset_train[feature]=dataset_train[feature].map(labels_ordered)
```

```python
feature_scale = [feature for feature in dataset_train.columns if feature not in ['Id', 'SalePrice'] ]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset_train[feature_scale])
```

```python
scaler.transform(dataset_train[feature_scale])
```

```python
data_train = pd.concat([dataset_train[['Id', 'SalePrice']].reset_index(drop = True), pd.DataFrame(scaler.transform(dataset_train[feature_scale]), columns = feature_scale)], axis = 1)
```

```python
data_train.head()
```

```python
data_train.to_csv('X_train.csv', index = False)
```
